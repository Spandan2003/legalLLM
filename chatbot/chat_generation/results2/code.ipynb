```python
import pandas as pd

df1 = pd.read_csv('./chatbot/chat_generation/results2/app_hall1.csv', index_col=0)
df1['Simulated Chat'] = df1['Simulated Chat'].apply(lambda x: ("Chatbot:" + x.split("Chatbot:", maxsplit=1)[1]).replace('Chatbot: ', 'AI: ').replace('Simulated User: ', 'Human: ')[:-15])
df1["Chatbot_version"] = "v1"
df2 = pd.read_csv('./chatbot/chat_generation/results2/app_hall2.csv', index_col=0)
df2['Simulated Chat'] = df2['Simulated Chat'].apply(lambda x: ("Chatbot:" + x.split("Chatbot:", maxsplit=1)[1]).replace('Chatbot: ', 'AI: ').replace('Simulated User: ', 'Human: ')[:-15])
df2["Chatbot_version"] = "v2"
df3 = pd.read_csv('./chatbot/chat_generation/results2/app_hall3.csv', index_col=0)
df3['Simulated Chat'] = df3['Simulated Chat'].apply(lambda x: ("Chatbot:" + x.split("Chatbot:", maxsplit=1)[1]).replace('Chatbot: ', 'AI: ').replace('Simulated User: ', 'Human: ')[:-15])
df3["Chatbot_version"] = "v3"
df4 = pd.read_csv('./chatbot/chat_generation/results2/app_hall4.csv', index_col=0)
df4['Simulated Chat'] = df4['Simulated Chat'].apply(lambda x: ("Chatbot:" + x.split("Chatbot:", maxsplit=1)[1]).replace('Chatbot: ', 'AI: ').replace('Simulated User: ', 'Human: ')[:-15])
df4["Chatbot_version"] = "v4"
```

```python
combined_df = pd.concat([df1, df2, df3, df4])
combined_df.reset_index(drop=True, inplace=True)
combined_df.to_csv('./chatbot/chat_generation/results2/combined.csv', index=False)
combined_df
```
