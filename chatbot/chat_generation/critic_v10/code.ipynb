```python
import pandas as pd
```

```python
data = pd.read_csv('./chatbot/chat_generation/critic_v10/result_df_v1 copy.csv', index_col=0)
data
```

```python
data['inconsistency_count'] = data['Result'].str.count(r'\b\d+\.')
data['turn_count'] = data['Chat'].str.count(r'AI:') - 1
mean_values = data.groupby('Version').apply(lambda x: (x['inconsistency_count'].sum() / x['turn_count'].sum()))
display(mean_values)
mean_values = data.groupby('Version').apply(lambda x: (x['inconsistency_count'] / x['turn_count']).mean())
display(mean_values)
data
```
