```python
import tkinter as tk
from tkinter import scrolledtext, messagebox

# Example of multiple data points
data_points = [
    {
        'context': '\nDocument 0:\n3. The next step is to reach the Medi',
        'chat_history': '\ufeffChatbot: Hi! I am your consumer grievance assista',
        'history': 'The user shares that they got a tattoo from XYZ Ta',
        'query': 'I need them to cover my medical expenses',
        'reformulated': 'The user is seeking compensation for the medical e',
        'response': 'Covering medical expenses is a reasonable expectat',
        'analysis_temp': 'Inconsistencies Present: Yes\n\nInconsistencies:\n1. ',
        'analysis': 'Inconsistency Present: Yes\nExplanation for Correct',
        'modified_analysis': ''
    },
    {
        'context': '\nDocument 1:\n5. User is requesting a refund from ABC Corp.',
        'chat_history': '\ufeffChatbot: Hello! How can I help you today?',
        'history': 'User complains about an issue with a purchase',
        'query': 'I want my money back!',
        'reformulated': 'The user is requesting a refund for a defective product.',
        'response': 'Refunds are possible if the product is returned within 30 days.',
        'analysis_temp': 'Inconsistencies Present: No',
        'analysis': 'No inconsistency detected.',
        'modified_analysis': ''
    }
]

class AnnotatorApp:
    def __init__(self, root, data_points):
        self.root = root
        self.data_points = data_points
        self.current_index = 0  # Start from the first data point
        self.evaluation_result = None  # To store the evaluation (Yes/No)
        
        self.root.title("Annotator Evaluation System")
        self.root.geometry("800x600")
        
        # Create UI components
        self.create_widgets()

    def create_widgets(self):
        # Title Label
        title_label = tk.Label(self.root, text="Annotator Evaluation System", font=("Helvetica", 16))
        title_label.pack(pady=10)

        # Chat History Label and Text box
        self.chat_history_label = tk.Label(self.root, text="Chat History", font=("Helvetica", 12))
        self.chat_history_label.pack(anchor="w", padx=10)
        
        self.chat_history_text = tk.Label(self.root, text="", anchor="w", justify="left", wraplength=750)
        self.chat_history_text.pack(fill=tk.X, padx=10, pady=5)

        # Context Label and Scrolled Text box
        self.context_label = tk.Label(self.root, text="Context", font=("Helvetica", 12))
        self.context_label.pack(anchor="w", padx=10)

        self.context_text = scrolledtext.ScrolledText(self.root, wrap=tk.WORD, width=90, height=10)
        self.context_text.pack(padx=10, pady=5)
        self.context_text.config(state=tk.DISABLED)

        # Modified Analysis Label and Text box
        self.modified_analysis_label = tk.Label(self.root, text="Modified Analysis", font=("Helvetica", 12))
        self.modified_analysis_label.pack(anchor="w", padx=10)

        self.modified_analysis_text = scrolledtext.ScrolledText(self.root, wrap=tk.WORD, width=90, height=10)
        self.modified_analysis_text.pack(padx=10, pady=5)

        # Navigation buttons
        self.navigation_frame = tk.Frame(self.root)
        self.navigation_frame.pack(pady=10)

        self.prev_button = tk.Button(self.navigation_frame, text="Previous", command=self.prev_data_point)
        self.prev_button.grid(row=0, column=0, padx=5)

        self.next_button = tk.Button(self.navigation_frame, text="Next", command=self.next_data_point)
        self.next_button.grid(row=0, column=1, padx=5)

        # Evaluation (Yes/No) Buttons
        self.evaluation_frame = tk.Frame(self.root)
        self.evaluation_frame.pack(pady=10)

        self.yes_button = tk.Button(self.evaluation_frame, text="Yes", command=lambda: self.save_evaluation("Yes"))
        self.yes_button.grid(row=0, column=0, padx=5)

        self.no_button = tk.Button(self.evaluation_frame, text="No", command=lambda: self.save_evaluation("No"))
        self.no_button.grid(row=0, column=1, padx=5)

        # Load the first data point
        self.load_data_point()

    def load_data_point(self):
        # Load the current data point
        data_point = self.data_points[self.current_index]

        # Update Chat History
        self.chat_history_text.config(text=data_point['chat_history'])

        # Update Context (Scrollable Text)
        self.context_text.config(state=tk.NORMAL)
        self.context_text.delete("1.0", tk.END)
        self.context_text.insert(tk.END, data_point['context'])
        self.context_text.config(state=tk.DISABLED)

        # Update Modified Analysis (Editable Text)
        self.modified_analysis_text.delete("1.0", tk.END)
        self.modified_analysis_text.insert(tk.END, data_point['modified_analysis'])

    def prev_data_point(self):
        if self.current_index > 0:
            self.current_index -= 1
            self.load_data_point()

    def next_data_point(self):
        if self.current_index < len(self.data_points) - 1:
            self.current_index += 1
            self.load_data_point()

    def save_evaluation(self, evaluation):
        # Save the evaluation for the current data point
        self.evaluation_result = evaluation
        data_point = self.data_points[self.current_index]
        data_point['evaluation'] = self.evaluation_result  # Save evaluation in the data point
        messagebox.showinfo("Success", f"Evaluation ({evaluation}) saved successfully!")

        # Optionally, move to the next data point
        self.next_data_point()

# Set up the main window
root = tk.Tk()

# Initialize the app with the root window and the data points
app = AnnotatorApp(root, data_points)

# Run the GUI loop
root.mainloop()
```
