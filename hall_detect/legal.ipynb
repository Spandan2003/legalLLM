```python
import pandas as pd
import requests

# Function to convert URL to plain text format and fetch the content
def get_google_doc_text(doc_url):
    # Extract the document ID from the URL
    try:
        doc_id = doc_url.split('/d/')[1].split('/')[0]
        text_url = f"https://docs.google.com/document/d/{doc_id}/export?format=txt"
    except:
        print(f"Failed to retrieve document: {doc_url}")
        return None
    # Make a GET request to fetch the text content
    response = requests.get(text_url)
    
    # Check if the request was successful
    if response.status_code == 200:
        return response.text
    else:
        print(f"Failed to retrieve document: {doc_url}")
        return None

# Apply the function to each URL and save the results in a new DataFrame column
data1['Chat'] = data1['Chat Link'].apply(get_google_doc_text)
data2['Chat'] = data2['Chat Link'].apply(get_google_doc_text)
data1

# Display the DataFrame with the extracted text
display(data1)
display(data2)
data1.to_csv("./code/legalLLM/hall_detect/data1.csv")
data2.to_csv("./code/legalLLM/hall_detect/data2.csv")
```

```python
data = pd.read_csv("./code/legalLLM/hall_detect/data1_modified.csv", index_col=0)
data.dropna(axis=0, subset=["Chat"], inplace=True)
data.fillna("No", inplace=True)
data.reset_index(inplace=True, drop=True)
cols = ["Incorrect organisation", "Incorrect Legal Knowledge", "Incorrect Methodology of Grievance Redressal", "Incorrect private information", "Incorrect Logic", "Mistake in notice or complaint"]
for col in cols:
    data[col] = data[col].map({"Yes": 1, "No": 0})
data
```

```python
data_mod["hall"] = data[cols].sum(axis=1)
data_mod["hall"]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

data_mod = data[cols]

# 1. Overview of Data
print("Overview of Hallucinations")
display(data_mod.describe())  # Summary statistics
print("\nHallucination Presence Count")
display(data_mod.sum())  # Total occurrences of each hallucination

# 2. Hallucination Count Analysis
hallucination_counts = data_mod.sum().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
sns.barplot(x=hallucination_counts.index, y=hallucination_counts.values, palette="viridis")
plt.xticks(rotation=45, ha="right")
plt.title("Hallucination Count Analysis")
plt.ylabel("Count")
plt.xlabel("Hallucination Type")
plt.show()

# 3. Correlation Analysis
plt.figure(figsize=(10, 6))
correlation_matrix = data_mod.corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Between Hallucination Types")
plt.show()

# 4. Pair Analysis (Hallucinations Occurring Together)
pair_counts = data_mod.T.dot(data_mod)  # Cross-product counts pairs of hallucinations
plt.figure(figsize=(10, 6))
sns.heatmap(pair_counts, annot=True, fmt="d", cmap="Blues")
plt.title("Pairwise Occurrence of Hallucinations")
plt.show()
```
